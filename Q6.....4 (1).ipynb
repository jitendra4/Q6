```python
driver.get('https://www.billboard.com/')
```

```python
#clickng on top 10 option 
top_10=driver.find_element_by_xpath("//a[@class='pill-button text--left display--inline-block shadow--soft']")
top_10.click()
```

```python
#intialising lists
song = []
artist = []
peak_rank = []
peak_date = []
```

```python

#Songs
songs=driver.find_elements_by_xpath("//div[@class='ye-chart-item__title']")
for i in songs:
    song.append(i.text)
```

```python

#artist
artists=driver.find_elements_by_xpath("//div[@class='ye-chart-item__artist']")
for i in artists:
    artist.append(i.text)
```

```python
#peak_rank
peak_ranks=driver.find_elements_by_xpath("//span[@class='decade-end-chart-item__peak-info-rank']")
for i in peak_ranks:
    peak_rank.append(i.text)
    
print(peak_rank)
```

```python
#peak_rank
peak_dates=driver.find_elements_by_xpath("//span[@class='decade-end-chart-item__peak-info-data']")
for i in peak_dates:
    peak_date.append(i.text)
    
print(peak_date)
```

```python
df=pd.DataFrame({})
df["songs"]=song[0:20]
df["artists"]=artist[0:20]
df["peak_ranks"]=peak_rank[0:20]
df["peak_ranks"]=peak_rank[0:20]
df["peak_dates"]=peak_date[0:20]
df
```
